```python
from ultralytics import YOLO
import os
import torch
import numpy as np
import cv2
from collections import defaultdict # (ìˆ˜ì •) ì¢…ë¥˜ë³„ë¡œ ì¹´ìš´íŠ¸í•˜ê¸° ìœ„í•´ import

# --- 1. ëª¨ë¸ ë¡œë”© ---
HOME = '.'
model_path = os.path.join(HOME, 'runs/detect/train2/weights/best.pt')

if not os.path.exists(model_path):
    print(f"ì˜¤ë¥˜: ëª¨ë¸ íŒŒì¼ '{model_path}'ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ì„ì‹œë¡œ 'yolov8n.pt'ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.")
    model = YOLO('yolov8n.pt')
else:
    model = YOLO(model_path)
    print(f"'{model_path}'ì—ì„œ ëª¨ë¸ì„ ì„±ê³µì ìœ¼ë¡œ ë¶ˆëŸ¬ì™”ìŠµë‹ˆë‹¤.")


# --- 2. IoU ê³„ì‚° í•¨ìˆ˜ (ë³€ê²½ ì—†ìŒ) ---
def calculate_iou(box_a, box_b):
    xA = max(box_a[0], box_b[0])
    yA = max(box_a[1], box_b[1])
    xB = min(box_a[2], box_b[2])
    yB = min(box_a[3], box_b[3])
    inter_area = max(0, xB - xA) * max(0, yB - yA)
    box_a_area = (box_a[2] - box_a[0]) * (box_a[3] - box_a[1])
    box_b_area = (box_b[2] - box_b[0]) * (box_b[3] - box_b[1])
    union_area = box_a_area + box_b_area - inter_area
    iou = inter_area / union_area
    return iou


# --- 3. ì„œë¡œ ë‹¤ë¥¸ 3ê°œì˜ ì´ë¯¸ì§€ì—ì„œ ì°¨ëŸ‰ ë° ì¢…ë¥˜ íƒì§€ ---
image_files = ["park.jpg", "park.jpg", "park.jpg"]
all_detections = []
class_names = [] # (ìˆ˜ì •) í´ë˜ìŠ¤ ì´ë¦„ ëª©ë¡ì„ ì €ì¥í•  ë³€ìˆ˜

for i, img_file in enumerate(image_files):
    if not os.path.exists(img_file):
        print(f"ì˜¤ë¥˜: ì´ë¯¸ì§€ íŒŒì¼ '{img_file}'ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ê±´ë„ˆëœë‹ˆë‹¤.")
        continue

    print(f"'{img_file}'ì—ì„œ ì°¨ëŸ‰ì„ íƒì§€í•©ë‹ˆë‹¤...")
    results = model.predict(img_file, conf=0.5)

    # (ìˆ˜ì • 1) í´ë˜ìŠ¤ ì´ë¦„ì„ ëª¨ë¸ì—ì„œ ê°€ì ¸ì˜¤ê¸° (í•œ ë²ˆë§Œ ìˆ˜í–‰)
    if i == 0:
        class_names = results[0].names
        print("íƒì§€ ê°€ëŠ¥í•œ í´ë˜ìŠ¤:", class_names)

    # (ìˆ˜ì • 2) ë°”ìš´ë”© ë°•ìŠ¤ì™€ í´ë˜ìŠ¤ IDë¥¼ í•¨ê»˜ ì €ì¥
    detections_data = []
    boxes_tensor = results[0].boxes.xyxy.cpu()
    classes_tensor = results[0].boxes.cls.cpu()

    for box, cls_id in zip(boxes_tensor, classes_tensor):
        detections_data.append({
            'box': box.numpy().tolist(),  # ë°•ìŠ¤ ì¢Œí‘œ
            'class_id': int(cls_id.item()) # í´ë˜ìŠ¤ ID
        })

    all_detections.append(detections_data)
    print(f"íƒì§€ëœ ì°¨ëŸ‰ ìˆ˜: {len(detections_data)}")


# --- 4. ìœ„ì¹˜ ë° ì¢…ë¥˜ ë¹„êµë¥¼ í†µí•œ ë¶ˆë²• ì£¼ì •ì°¨ ì°¨ëŸ‰ ì‹ë³„ ---
if len(all_detections) == 3 and class_names:
    detections_1, detections_2, detections_3 = all_detections
    IOU_THRESHOLD = 0.9

    # 1ì°¨ ë¹„êµ: ê²°ê³¼ë¬¼ì— í´ë˜ìŠ¤ ì •ë³´ë„ í¬í•¨
    stationary_after_5_mins = []
    for det1 in detections_1:
        for det2 in detections_2:
            # (ìˆ˜ì • 3) ë°•ìŠ¤ ì¢Œí‘œë¡œ IoU ê³„ì‚°
            if calculate_iou(det1['box'], det2['box']) > IOU_THRESHOLD:
                stationary_after_5_mins.append(det1) # boxì™€ class_idê°€ ëª¨ë‘ ë‹´ê¸´ ë”•ì…”ë„ˆë¦¬ë¥¼ ì¶”ê°€
                break

    # 2ì°¨ ë¹„êµ: ìµœì¢… ë¶ˆë²• ì£¼ì°¨ ì°¨ëŸ‰ í™•ì •
    illegally_parked_cars = []
    for det1_2 in stationary_after_5_mins:
        for det3 in detections_3:
            if calculate_iou(det1_2['box'], det3['box']) > IOU_THRESHOLD:
                illegally_parked_cars.append(det1_2)
                break

    # --- 5. (ìˆ˜ì •) ìµœì¢… ê²°ê³¼ ì§‘ê³„ ë° ì¶œë ¥ ---
    print(f"\nìµœì¢…ì ìœ¼ë¡œ 10ë¶„ ì´ìƒ ë¶ˆë²• ì£¼ì°¨ë¡œ ì˜ì‹¬ë˜ëŠ” ì°¨ëŸ‰ì€ ì´ {len(illegally_parked_cars)}ëŒ€ ì…ë‹ˆë‹¤.")

    # ì°¨ì¢…ë³„ë¡œ ì¹´ìš´íŠ¸
    parking_summary = defaultdict(int)
    for car in illegally_parked_cars:
        car_class_id = car['class_id']
        car_class_name = class_names[car_class_id] # IDë¥¼ ì´ë¦„ìœ¼ë¡œ ë³€í™˜
        parking_summary[car_class_name] += 1

    if parking_summary:
        print("\n--- ì°¨ì¢…ë³„ ë¶ˆë²• ì£¼ì •ì°¨ ìš”ì•½ ---")
        for car_type, count in parking_summary.items():
            print(f"- {car_type}: {count}ëŒ€")
        print("---------------------------------")


    # --- 6. (ìˆ˜ì •) ê²°ê³¼ ì‹œê°í™” ê°œì„  ---
    image_to_draw = cv2.imread(image_files[0])


    for car in illegally_parked_cars:
        box = car['box']
        class_id = car['class_id']
        class_name = class_names[class_id] # IDë¥¼ ì´ë¦„ìœ¼ë¡œ ë³€í™˜

        x1, y1, x2, y2 = map(int, box)

        # ì‹œê°í™”
        cv2.rectangle(image_to_draw, (x1, y1), (x2, y2), (0, 0, 255), 3)

        # í‘œì‹œí•  í…ìŠ¤íŠ¸: "Illegal Parking: truck" ê³¼ ê°™ì€ í˜•íƒœ
        label = f"Illegal Parking: {class_name}"
        cv2.putText(image_to_draw, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 0, 255), 2)

    cv2.imwrite("result_parking_with_type.jpg", image_to_draw)
    print("\nì°¨ëŸ‰ ì¢…ë¥˜ê°€ í‘œì‹œëœ ê²°ê³¼ê°€ 'result_parking_with_type.jpg' íŒŒì¼ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")

else:
    print("\nì´ë¯¸ì§€ íŒŒì¼ì´ 3ê°œê°€ ì•„ë‹ˆê±°ë‚˜ í´ë˜ìŠ¤ ì´ë¦„ì„ ë¶ˆëŸ¬ì˜¬ ìˆ˜ ì—†ì–´ ë¹„êµë¥¼ ìˆ˜í–‰í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
```

```python
import cv2
from ultralytics import YOLO

# 1. ëª¨ë¸ ë¡œë“œ
# ì§ì ‘ í•™ìŠµì‹œí‚¨ YOLO ëª¨ë¸ì˜ ê²½ë¡œë¥¼ ì§€ì •í•©ë‹ˆë‹¤. (ì˜ˆ: 'best.pt', 'my_model.pt')
try:
    model = YOLO('your_custom_model.pt')
except Exception as e:
    print(f"Error loading model: {e}")
    print("ëª¨ë¸ íŒŒì¼('your_custom_model.pt')ì´ í˜„ì¬ í´ë”ì— ìˆëŠ”ì§€ í™•ì¸í•´ì£¼ì„¸ìš”.")
    exit()

# 2. ìŠ¤ë§ˆíŠ¸í° ì¹´ë©”ë¼ ì˜ìƒ ìŠ¤íŠ¸ë¦¼ ì£¼ì†Œ
# 'IP Webcam' ì•±ì— í‘œì‹œëœ ì£¼ì†Œ ë’¤ì— '/video'ë¥¼ ì¶”ê°€í•©ë‹ˆë‹¤.
# !!! ì•„ë˜ URLì„ ìì‹ ì˜ í™˜ê²½ì— ë§ê²Œ ìˆ˜ì •í•´ì£¼ì„¸ìš” !!!
video_url = "http://192.168.1.5:8080/video"

# ë¹„ë””ì˜¤ ìº¡ì²˜ ê°ì²´ ìƒì„±
cap = cv2.VideoCapture(video_url)

if not cap.isOpened():
    print(f"ì˜¤ë¥˜: ë¹„ë””ì˜¤ ìŠ¤íŠ¸ë¦¼ì„ ì—´ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. URLì„ í™•ì¸í•˜ì„¸ìš”: {video_url}")
    exit()
else:
    print("ë¹„ë””ì˜¤ ìŠ¤íŠ¸ë¦¼ì— ì„±ê³µì ìœ¼ë¡œ ì—°ê²°ë˜ì—ˆìŠµë‹ˆë‹¤.")

# 3. ì‹¤ì‹œê°„ ì˜ìƒ ì²˜ë¦¬ ë£¨í”„
while True:
    # í”„ë ˆì„ ë‹¨ìœ„ë¡œ ì˜ìƒ ì½ê¸°
    success, frame = cap.read()

    if not success:
        print("í”„ë ˆì„ì„ ë°›ì•„ì˜¤ì§€ ëª»í–ˆìŠµë‹ˆë‹¤. ì—°ê²°ì„ í™•ì¸í•©ë‹ˆë‹¤.")
        break

    # 4. ëª¨ë¸ì„ ì‚¬ìš©í•˜ì—¬ ê°ì²´ íƒì§€ ìˆ˜í–‰
    # stream=True ì˜µì…˜ì€ ëŒ€ìš©ëŸ‰ ì˜ìƒ ì²˜ë¦¬ì— ë” íš¨ìœ¨ì ì…ë‹ˆë‹¤.
    results = model(frame, stream=True)

    # íƒì§€ëœ ê°ì²´ë“¤ì— ëŒ€í•œ ì •ë³´ ì²˜ë¦¬
    for r in results:
        boxes = r.boxes
        for box in boxes:
            # ë°”ìš´ë”© ë°•ìŠ¤ ì¢Œí‘œ (x1, y1, x2, y2)
            x1, y1, x2, y2 = map(int, box.xyxy[0])

            # í´ë˜ìŠ¤ ID ë° ì´ë¦„
            cls = int(box.cls[0])
            class_name = model.names[cls]

            # ì‹ ë¢°ë„ ì ìˆ˜
            conf = float(box.conf[0])

            # 5. í™”ë©´ì— ë°”ìš´ë”© ë°•ìŠ¤ì™€ ë¼ë²¨ í‘œì‹œ
            # ì‹ ë¢°ë„ê°€ 0.5 ì´ìƒì¸ ê²½ìš°ì—ë§Œ í‘œì‹œ (í•„ìš”ì— ë”°ë¼ ì¡°ì ˆ)
            if conf > 0.5:
                # ë°”ìš´ë”© ë°•ìŠ¤ ê·¸ë¦¬ê¸°
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

                # ë¼ë²¨ í…ìŠ¤íŠ¸ (í´ë˜ìŠ¤ ì´ë¦„ + ì‹ ë¢°ë„)
                label = f'{class_name} {conf:.2f}'

                # í…ìŠ¤íŠ¸ ë°°ê²½ ê·¸ë¦¬ê¸°
                (w, h), _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 2)
                cv2.rectangle(frame, (x1, y1 - h - 10), (x1 + w, y1), (0, 255, 0), -1)

                # í…ìŠ¤íŠ¸ ì“°ê¸°
                cv2.putText(frame, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)


    # ê²°ê³¼ ì˜ìƒ ì¶œë ¥
    cv2.imshow("Smartphone Object Detection", frame)

    # 'q' í‚¤ë¥¼ ëˆ„ë¥´ë©´ ë£¨í”„ ì¢…ë£Œ
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

# 6. ìì› í•´ì œ
cap.release()
cv2.destroyAllWindows()
print("í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.")
```

```python
import numpy as np
import supervision as sv
from ultralytics import YOLO
bestFile = "/content/runs/detect/train2/weights/best.pt"
model = YOLO(bestFile)
box_annotator = sv.BoxAnnotator()

def callback(frame: np.ndarray, _: int) -> np.ndarray:
    results = model(frame)[0]
    detections = sv.Detections.from_ultralytics(results)
    return box_annotator.annotate(frame.copy(), detections=detections)

sv.process_video(
    source_path="CCTV_load.mp4",
    target_path="result.mp4",
    callback=callback
)
```
