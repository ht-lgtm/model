```python
from ultralytics import YOLO
import os
import torch
import numpy as np
import cv2
from collections import defaultdict # (수정) 종류별로 카운트하기 위해 import

# --- 1. 모델 로딩 ---
HOME = '.'
model_path = os.path.join(HOME, 'runs/detect/train2/weights/best.pt')

if not os.path.exists(model_path):
    print(f"오류: 모델 파일 '{model_path}'를 찾을 수 없습니다. 임시로 'yolov8n.pt'를 사용합니다.")
    model = YOLO('yolov8n.pt')
else:
    model = YOLO(model_path)
    print(f"'{model_path}'에서 모델을 성공적으로 불러왔습니다.")


# --- 2. IoU 계산 함수 (변경 없음) ---
def calculate_iou(box_a, box_b):
    xA = max(box_a[0], box_b[0])
    yA = max(box_a[1], box_b[1])
    xB = min(box_a[2], box_b[2])
    yB = min(box_a[3], box_b[3])
    inter_area = max(0, xB - xA) * max(0, yB - yA)
    box_a_area = (box_a[2] - box_a[0]) * (box_a[3] - box_a[1])
    box_b_area = (box_b[2] - box_b[0]) * (box_b[3] - box_b[1])
    union_area = box_a_area + box_b_area - inter_area
    iou = inter_area / union_area
    return iou


# --- 3. 서로 다른 3개의 이미지에서 차량 및 종류 탐지 ---
image_files = ["park.jpg", "park.jpg", "park.jpg"]
all_detections = []
class_names = [] # (수정) 클래스 이름 목록을 저장할 변수

for i, img_file in enumerate(image_files):
    if not os.path.exists(img_file):
        print(f"오류: 이미지 파일 '{img_file}'을 찾을 수 없습니다. 건너뜁니다.")
        continue

    print(f"'{img_file}'에서 차량을 탐지합니다...")
    results = model.predict(img_file, conf=0.5)

    # (수정 1) 클래스 이름을 모델에서 가져오기 (한 번만 수행)
    if i == 0:
        class_names = results[0].names
        print("탐지 가능한 클래스:", class_names)

    # (수정 2) 바운딩 박스와 클래스 ID를 함께 저장
    detections_data = []
    boxes_tensor = results[0].boxes.xyxy.cpu()
    classes_tensor = results[0].boxes.cls.cpu()

    for box, cls_id in zip(boxes_tensor, classes_tensor):
        detections_data.append({
            'box': box.numpy().tolist(),  # 박스 좌표
            'class_id': int(cls_id.item()) # 클래스 ID
        })

    all_detections.append(detections_data)
    print(f"탐지된 차량 수: {len(detections_data)}")


# --- 4. 위치 및 종류 비교를 통한 불법 주정차 차량 식별 ---
if len(all_detections) == 3 and class_names:
    detections_1, detections_2, detections_3 = all_detections
    IOU_THRESHOLD = 0.9

    # 1차 비교: 결과물에 클래스 정보도 포함
    stationary_after_5_mins = []
    for det1 in detections_1:
        for det2 in detections_2:
            # (수정 3) 박스 좌표로 IoU 계산
            if calculate_iou(det1['box'], det2['box']) > IOU_THRESHOLD:
                stationary_after_5_mins.append(det1) # box와 class_id가 모두 담긴 딕셔너리를 추가
                break

    # 2차 비교: 최종 불법 주차 차량 확정
    illegally_parked_cars = []
    for det1_2 in stationary_after_5_mins:
        for det3 in detections_3:
            if calculate_iou(det1_2['box'], det3['box']) > IOU_THRESHOLD:
                illegally_parked_cars.append(det1_2)
                break

    # --- 5. (수정) 최종 결과 집계 및 출력 ---
    print(f"\n최종적으로 10분 이상 불법 주차로 의심되는 차량은 총 {len(illegally_parked_cars)}대 입니다.")

    # 차종별로 카운트
    parking_summary = defaultdict(int)
    for car in illegally_parked_cars:
        car_class_id = car['class_id']
        car_class_name = class_names[car_class_id] # ID를 이름으로 변환
        parking_summary[car_class_name] += 1

    if parking_summary:
        print("\n--- 차종별 불법 주정차 요약 ---")
        for car_type, count in parking_summary.items():
            print(f"- {car_type}: {count}대")
        print("---------------------------------")


    # --- 6. (수정) 결과 시각화 개선 ---
    image_to_draw = cv2.imread(image_files[0])


    for car in illegally_parked_cars:
        box = car['box']
        class_id = car['class_id']
        class_name = class_names[class_id] # ID를 이름으로 변환

        x1, y1, x2, y2 = map(int, box)

        # 시각화
        cv2.rectangle(image_to_draw, (x1, y1), (x2, y2), (0, 0, 255), 3)

        # 표시할 텍스트: "Illegal Parking: truck" 과 같은 형태
        label = f"Illegal Parking: {class_name}"
        cv2.putText(image_to_draw, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 0, 255), 2)

    cv2.imwrite("result_parking_with_type.jpg", image_to_draw)
    print("\n차량 종류가 표시된 결과가 'result_parking_with_type.jpg' 파일로 저장되었습니다.")

else:
    print("\n이미지 파일이 3개가 아니거나 클래스 이름을 불러올 수 없어 비교를 수행할 수 없습니다.")
```

```python
import cv2
from ultralytics import YOLO

# 1. 모델 로드
# 직접 학습시킨 YOLO 모델의 경로를 지정합니다. (예: 'best.pt', 'my_model.pt')
try:
    model = YOLO('your_custom_model.pt')
except Exception as e:
    print(f"Error loading model: {e}")
    print("모델 파일('your_custom_model.pt')이 현재 폴더에 있는지 확인해주세요.")
    exit()

# 2. 스마트폰 카메라 영상 스트림 주소
# 'IP Webcam' 앱에 표시된 주소 뒤에 '/video'를 추가합니다.
# !!! 아래 URL을 자신의 환경에 맞게 수정해주세요 !!!
video_url = "http://192.168.1.5:8080/video"

# 비디오 캡처 객체 생성
cap = cv2.VideoCapture(video_url)

if not cap.isOpened():
    print(f"오류: 비디오 스트림을 열 수 없습니다. URL을 확인하세요: {video_url}")
    exit()
else:
    print("비디오 스트림에 성공적으로 연결되었습니다.")

# 3. 실시간 영상 처리 루프
while True:
    # 프레임 단위로 영상 읽기
    success, frame = cap.read()

    if not success:
        print("프레임을 받아오지 못했습니다. 연결을 확인합니다.")
        break

    # 4. 모델을 사용하여 객체 탐지 수행
    # stream=True 옵션은 대용량 영상 처리에 더 효율적입니다.
    results = model(frame, stream=True)

    # 탐지된 객체들에 대한 정보 처리
    for r in results:
        boxes = r.boxes
        for box in boxes:
            # 바운딩 박스 좌표 (x1, y1, x2, y2)
            x1, y1, x2, y2 = map(int, box.xyxy[0])

            # 클래스 ID 및 이름
            cls = int(box.cls[0])
            class_name = model.names[cls]

            # 신뢰도 점수
            conf = float(box.conf[0])

            # 5. 화면에 바운딩 박스와 라벨 표시
            # 신뢰도가 0.5 이상인 경우에만 표시 (필요에 따라 조절)
            if conf > 0.5:
                # 바운딩 박스 그리기
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

                # 라벨 텍스트 (클래스 이름 + 신뢰도)
                label = f'{class_name} {conf:.2f}'

                # 텍스트 배경 그리기
                (w, h), _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 2)
                cv2.rectangle(frame, (x1, y1 - h - 10), (x1 + w, y1), (0, 255, 0), -1)

                # 텍스트 쓰기
                cv2.putText(frame, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)


    # 결과 영상 출력
    cv2.imshow("Smartphone Object Detection", frame)

    # 'q' 키를 누르면 루프 종료
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

# 6. 자원 해제
cap.release()
cv2.destroyAllWindows()
print("프로그램을 종료합니다.")
```

```python
import numpy as np
import supervision as sv
from ultralytics import YOLO
bestFile = "/content/runs/detect/train2/weights/best.pt"
model = YOLO(bestFile)
box_annotator = sv.BoxAnnotator()

def callback(frame: np.ndarray, _: int) -> np.ndarray:
    results = model(frame)[0]
    detections = sv.Detections.from_ultralytics(results)
    return box_annotator.annotate(frame.copy(), detections=detections)

sv.process_video(
    source_path="CCTV_load.mp4",
    target_path="result.mp4",
    callback=callback
)
```
